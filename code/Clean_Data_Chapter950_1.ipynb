```python
import xml.etree.ElementTree as ET
import time
import pandas as pd
```

```python
noparking = ET.parse('Ch_950_Sch_13_NoParking.xml')
root = noparking.getroot()
head = []
```

```python
def getvalueofnode(node):
    """ return node text or None """
    return node.text if node is not None else None
```

```python
dfcols = ['highway', 'side', 'between', 'time']
df_xml = pd.DataFrame(columns=dfcols)
for node in root:
    highway = node.find('Highway')
    side = node.find('Side')
    between = node.find('Between')
    time = node.find('Prohibited_Times_and_or_Days')
    
    df_xml = df_xml.append(pd.Series([getvalueofnode(highway), getvalueofnode(side),
                       getvalueofnode(between),getvalueofnode(time)], index=dfcols),
            ignore_index=True)
```

```python
df_xml
```

```python
df_xml['start_loc'] = df_xml.between.str.split("and").str[0]
df_xml['end_loc'] = df_xml.between.str.split("and").str[1]
df_xml
```

```python
df = df_xml.dropna()
```
