```python
df_geotab = pd.read_csv('../data/cleaned/SearchingForParking_Toronto.csv')
```

```python
df_merge = pd.merge(df_geotab, total_parking, how='outer', on='Geohash')[['Geohash', 'latitude', 'longitude', 'TotalSearching', 'total_spots']]
```

```python
df_merge.fillna(0, inplace=True)
```

```python
df_merge['geohash_simpler'] = df_merge['Geohash'].apply(lambda x: x[:5])
```

```python
df_merge.groupby(['geohash_simpler'])['latitude'].agg('max').reset_index()
```

```python
df_merge.groupby(['geohash_simpler'])['longitude'].agg('min').reset_index()
```

```python
df_ratio = df_merge.groupby(['geohash_simpler'])['total_spots'].agg('sum').reset_index()
```

```python
df_ratio['total_search'] = df_merge.groupby(['geohash_simpler'])['TotalSearching'].agg('sum').reset_index()['TotalSearching']
```

```python
df_ratio['latitude'] = [pgh.decode_exactly(gh)[0] for gh in df_ratio['geohash_simpler'].tolist()]
```

```python
df_ratio['longitude'] = [pgh.decode_exactly(gh)[1] for gh in df_ratio['geohash_simpler'].tolist()]
```

```python
df_ratio.head()
```

```python
df_ratio['total_spots'] = df_ratio['total_spots'].apply(lambda x: max(x, 1))
```

```python
df_ratio['ratio'] = df_ratio['total_search'].divide(df_ratio['total_spots'])
```

```python
df_ratio.head()
```

```python
#### Create heatmap data
data_ratio = df_ratio[['geohash_simpler','ratio','latitude','longitude']] 
data_ratio['Value'] = data_ratio.apply(lambda row: [row.latitude, row.longitude, row.ratio], axis = 1)
heatdata_ratio = data_ratio.Value.tolist()
```

```python
# Add markers to layer
for index, row in data_totalsearching.iterrows():
    c = folium.Marker([row['Latitude'],row['Longitude']], popup = 'Geohash: ' + row.Geohash)
    c.add_to(Markers)
Markers.add_to(heatmap)
```

```python
#### Plot heatmap based on the Searching for Parking Index
index_ts = folium.FeatureGroup(name='Geohash - TotalSearching')
HeatMap(data = heatdata_totalsearching, min_opacity = 0.3, max_zoom = 18, max_val = 500, radius = 25).add_to(index_ts)
index_ts.add_to(heatmap)
```

```python
#### Plot heatmap based on the Searching for Parking Index
index_cap = folium.FeatureGroup(name='Geohash - Available parking spots')
HeatMap(data = heatdata_capacity, min_opacity = 0.3, max_zoom = 18, max_val = 500,
        gradient={1: 'blue', 50: 'lime', 200: 'red'}, radius = 25).add_to(index_cap) 
index_cap.add_to(heatmap)
```
